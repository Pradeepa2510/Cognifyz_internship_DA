Task 2 - City Analysis

```python
# Top 10 Cities with Most Restaurants
city_count = df['City'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=city_count.index, y=city_count.values, color='lightcoral')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Rating by City
city_avg = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=city_avg.index, y=city_avg.values, color='mediumseagreen')
plt.title("Average Rating by City")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🟩 Level 3 - Task 1 (Restaurant Reviews using Votes as proxy)

# Average number of votes per restaurant
avg_votes = df['Votes'].mean()
print("Average Votes per Restaurant:", round(avg_votes,2))

# Distribution of votes (to see engagement level)
plt.figure(figsize=(7,5))
sns.histplot(df['Votes'], bins=30, color='skyblue')
plt.title("Distribution of Votes (as Review Proxy)")
plt.xlabel("Votes")
plt.ylabel("Number of Restaurants")
plt.show()
```

Task 3 – Price Range vs Online Delivery / Table Booking

```python
# Calculate average price range based on service availability
avg_price_online_yes = df.loc[df['Has Online delivery'] == 'Yes', 'Price range'].mean()
avg_price_online_no  = df.loc[df['Has Online delivery'] == 'No',  'Price range'].mean()

avg_price_table_yes  = df.loc[df['Has Table booking'] == 'Yes', 'Price range'].mean()
avg_price_table_no   = df.loc[df['Has Table booking'] == 'No',  'Price range'].mean()
```

```python
compare_df = pd.DataFrame({
    'Service Type': ['Online Delivery', 'Online Delivery', 'Table Booking', 'Table Booking'],
    'Availability': ['Yes', 'No', 'Yes', 'No'],
    'Average Price Range': [avg_price_online_yes, avg_price_online_no,
                            avg_price_table_yes, avg_price_table_no]
})
compare_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Service Type',
    y='Average Price Range',
    hue='Availability',
    data=compare_df,
    palette='viridis'
)
plt.title("Average Price Range by Service Type and Availability")
plt.xlabel("Service Type")
plt.ylabel("Average Price Range")
plt.legend(title='Availability', loc='upper right')
plt.show()
```
